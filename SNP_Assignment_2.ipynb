```python
#Opening the PDF file

from PyPDF2 import PdfReader

pdf = PdfReader("harrypotter.pdf")

# Checking total number of pages in the book
len(pdf.pages)
```

```python
# My Date of Birth is 19-05-2003

# For file1:
# Birth date = 19 : so I am extracting pages 19 to 28

# For file2:
# Birth year = 2003 : using 103 : so I am extracting pages 103 to 112

def extract_pages(start, end, output_file):
    text = ""

    for i in range(start-1, end):   # -1 because indexing starts from 0
        text += pdf.pages[i].extract_text()

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(text)

# Creating file1
extract_pages(19, 28, "file1.txt")

# Creating file2
extract_pages(103, 112, "file2.txt")

print("Pages extracted successfully")
```

```python
# Displaying first few lines to verify the text extraction

with open("file1.txt") as f:
    print(f.read(500))
```

```python
# Word count for file1 using MapReduce concept

from collections import Counter

with open("file1.txt") as f:
    text = f.read()

# Converting text into words
words = text.split()

# MAP phase : creating (word, 1)
mapped = [(word, 1) for word in words]

# REDUCE phase : counting total occurrences
word_count = Counter(dict(mapped))

print("Word Count Output:\n")

for word, count in word_count.items():
    print(word, count)
```

```python
# Finding non-English words (like spells, names, places)

from spellchecker import SpellChecker

spell = SpellChecker()

with open("file2.txt") as f:
    text2 = f.read()

words2 = text2.split()

non_english = []

for word in words2:
    if word.lower() not in spell:
        non_english.append(word)

non_english_count = Counter(non_english)

print("Non-English Words and their counts:\n")

for word, count in non_english_count.items():
    print(word, count)
```

```python
# Saving the outputs into text files

with open("wordcount_output.txt", "w") as f:
    for word, count in word_count.items():
        f.write(f"{word} {count}\n")

with open("non_english_output.txt", "w") as f:
    for word, count in non_english_count.items():
        f.write(f"{word} {count}\n")

print("Outputs saved")
```
